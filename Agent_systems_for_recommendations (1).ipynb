### Creating agent

```python
!pip install smolagents
```

```python
from smolagents import tool
```

```python
from smolagents import ToolCallingAgent, InferenceClientModel
```

```python
model_id = "Qwen/Qwen2.5-Coder-32B-Instruct"
```

```python
model = InferenceClientModel(model_id=model_id)
```

```python
@tool
def get_predictions(movie_title: str) -> list[str]:
    """
    Function takes a movie title and generates a list of movies that are similar to it based on content.

    Args:
      movie_title(str): Title of the movie we are looking for similar movies to.

    Returns:
      movies(list(str)): List of movies that are similar to the movie_title.

    """

    docs = llama_retriever.retrieve(movie_title)

    movies = []

    for doc in docs:
      movies.append(doc.metadata.get("title"))

    return movies

content_based_agent = ToolCallingAgent(
    tools=[get_predictions],
    model=model,
    max_steps=2,
    name="content_based_recommender_agent",
    description="Finds content based recommendations",
)
```

```python
def generate_response(query):

  System_Message = """
  You are an AI assistant designed to help users to find movies similar to searched movie titles. Your primary goal is to provide helpful, precise and clear responses.

  You have access to the following tools:
  Tool Name: get_predictions, Description: takes a movie title and generates a list of movies that are similar to it based on content. , Arguments: movie_title(str), Outputs: movies(list(str))
  """

  message = f"""{System_Message}

      You have a movie name provided by the user in {query}.

      You have to provide the similar movies from the list obtained using the tool.
      Follow the guidelines:
      1. Try to provide relevant/accurate numbers if available.
      2. You don’t have to necessarily use all the information in the dataframe. Only choose information that is relevant.
      """

  return content_based_agent.run(task=message)
```

```python
user_item_interactions = df_ratings.pivot_table(index='userId', columns='movieId', values='rating')
```

```python
user_item_interactions
```

```python

user_item_interactions.fillna(0, inplace = True)
```

```python

user_item_interactions.shape
```

```python
##### Item on Item based
```

```python
item_item_matrix = user_item_interactions.T

```

```python
from sklearn.decomposition import TruncatedSVD


SVD = TruncatedSVD(n_components=10)
decomposed_matrix_item_item = SVD.fit_transform(item_item_matrix)
decomposed_matrix_item_item.shape
```

```python
@tool
def get_item_on_item_prediction(movie_name: str) -> list[str]:
  """
    Function takes a movie title and generates a list of movies that are similar to it based on collaborative filtering.

    Args:
      movie_name(str): Title of the movie we are looking for similar movies to.

    Returns:
      recommendations(list(str)): List of movies that are similar to the movie_title.

    """
  movie_index = df_metadata[df_metadata['title'] == movie_name].index
  if len(movie_index) == 0:
    return "Movie not found in metadata."
  movie_index = movie_index[0]
  correlation_scores = decomposed_matrix_item_item[movie_index]
  # Get the indices of movies with correlation scores greater than 0.90
  recommended_movie_indices = [i for i, score in enumerate(correlation_scores) if score > 0.90]

  recommendations = []
  for index in recommended_movie_indices:
    movie_id = item_item_matrix.index[index]
    # Ensure the movie_id exists in the metadata before appending
    if movie_id in df_metadata["id"].values:
        recommendations.append(df_metadata[df_metadata["id"] == movie_id]["title"].iloc[0])


  # Remove the original movie from recommendations if present
  if movie_name in recommendations:
      recommendations.remove(movie_name)

  return recommendations
```

```python
get_item_on_item_prediction("Toy Story")
```

```python
model_collaborative = InferenceClientModel(model_id=model_id)
```

```python
collaborative_based_agent = ToolCallingAgent(
    tools=[get_item_on_item_prediction],
    model=model_collaborative,
    max_steps=2,
    name="collaborative_based_recommender_agent",
    description="Finds similar items based recommendations",
  )
```

```python
def generate_response_collaborative(query):
  System_Message = """
  You are an AI assistant designed to help users to find movies similar to searched movie titles. Your primary goal is to provide helpful, precise and clear responses.

  You have access to the following tools:
  Tool Name: get_item_on_item_prediction, Description: takes a movie title and generates a list of movies that are similar to it based on collaborative filtering. , Arguments: movie_name(str), Outputs: recommendations(list(str))
  """

  message = f"""{System_Message}

      You have a movie name provided by the user in {query}.

      You have to provide the similar movies from the list obtained using the tool.
      Follow the guidelines:
      1. Try to provide relevant/accurate numbers if available.
      2. You don’t have to necessarily use all the information in the dataframe. Only choose information that is relevant.
      """

  return collaborative_based_agent.run(task=message)
```

```python
### Managed Model
```

```python
from smolagents import CodeAgent
```

```python
model_managed=InferenceClientModel("deepseek-ai/DeepSeek-R1", provider="together", max_tokens=8096)
```

```python
@tool
def merge_recommendations(lists: list[list[str]], top_k: int = 5) -> list[str]:
    """
    Merges multiple recommendation lists and returns the top K most frequent items.

    Args:
      lists (list): Variable number of lists containing recommended items.
      top_k (int): The number of top frequent items to return.

    Returns:
      list: A list of the top K most frequent recommended items.
    """
    from collections import Counter
    all_items = sum(lists, [])
    count = Counter(all_items)
    return [item for item, _ in count.most_common(top_k)]
```

```python
manager_agent = CodeAgent(
    name="ManagedRecommendationAgent",
    description="Coordinates three agents to produce the best recommendations.",
    model=model_managed,
    tools=[merge_recommendations],  # optional utility
    managed_agents=[
        content_based_agent,
        collaborative_based_agent,
        web_agent,
    ],
    additional_authorized_imports=[
        "pandas",
        "numpy",
        "collections",
        "json"
    ],
    planning_interval=3,
    verbosity_level=2,
    max_steps=3,
)
```
